```python
def check_nan(film: pd.DataFrame) -> None:
    
    nan_cols = film.isna().mean() * 100  
    
    display(f'N nan cols: {len(nan_cols[nan_cols>0])}')
    display(nan_cols[nan_cols>0])
    
    plt.figure(figsize=(10, 6))  

    sns.heatmap(film.isna(),  
                yticklabels=False,
                cmap='viridis',
                cbar=False)

    plt.show();
```

```python
check_nan(film)
```
